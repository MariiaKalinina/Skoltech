```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logistic_map(x, r):
    return r * x * (1. - x)
```

```python
def bifurcation_diagram(x0, r_min, r_max, num_steps, n, k):
    r_list = np.linspace(r_min, r_max, num_steps)
    bifurcations = []

    for r in r_list:
        x = x0
        for i in range(n + k):
            if i >= k:
                bifurcations.append([r, x])
            x = logistic_map(x, r)

    bifurcations = np.array(bifurcations)
    plt.plot(bifurcations[:,0], bifurcations[:,1], 'o', markersize=0.2)
    plt.xlabel('r')
    plt.ylim(0, 1)
    plt.ylabel('x')
    plt.show()
```

```python
bifurcation_diagram(0.1, 1, 5.0, 10000, 100, 50)
```
